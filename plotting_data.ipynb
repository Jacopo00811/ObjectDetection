```python
import pandas as pd
import plotly.express as px

# Read the data from the CSV file
df = pd.read_csv('timing_results.csv')

# Create a Plotly figure with facets for each image
fig = px.line(df, x='Num Boxes', y='Time (s)', color='Metric', markers=True, 
              facet_col='Image', title='Timing Comparison by Number of Boxes for Different Metrics')

# Show the figure
fig.show()
```
